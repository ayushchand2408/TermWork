```python
from sklearn.impute import KNNImputer
# a) Impute using the mean
df_mean_imputed = df.fillna(df.mean(numeric_only=True))
print("\n--- DataFrame with Nulls Replaced by Mean ---")
print(df_mean_imputed)
```

```python
# b) Impute using the median
df_median_imputed = df.fillna(df.median(numeric_only=True))
print("\n--- DataFrame with Nulls Replaced by Median ---")
print(df_median_imputed)

```

```python
# c) Impute using KNN Imputer
# This is a more advanced method that uses the average of the k-nearest neighbors
# to fill in the missing values.
df_numerical = df[['Age', 'Salary']]
imputer = KNNImputer(n_neighbors=2)  # We are looking at the 2 nearest neighbors.
df_knn_imputed = pd.DataFrame(imputer.fit_transform(df_numerical), columns=df_numerical.columns)
print(df_knn_imputed)
```

```python
import numpy as np
import pandas as pd

# The 'bins' list must be sorted in increasing or decreasing order.
# Here, it is correctly sorted in increasing order.
data = [2,7,10,13,19,24,34,56,67]
bins_list = [0,20,40,60,80]

# Digitize the data, assigning each value to its correct bin index
indices = np.digitize(data, bins_list)

# Create a DataFrame to show the values and their corresponding bin indices
df4 = pd.DataFrame({'Values': data, 'Bin_index': indices})

# Print the resulting DataFrame
print(df4)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Create a sample Titanic-like dataset from scratch
data = {
    'age': [22, 38, 26, 35, 35, np.nan, 54, 2, 27, 4],
    'fare': [7.25, 71.28, 7.92, 53.10, 8.05, np.nan, 51.86, 21.07, 13.00, 15.24],
    'class': ['Third', 'First', 'Third', 'First', 'Third', 'First', 'First', 'Third', 'Second', 'Third']
}
titanic = pd.DataFrame(data)

# Drop rows with any missing values in the 'age' or 'fare' columns
titanic = titanic.dropna(subset=['age', 'fare'])

# Create four age bins and label them
titanic['age_bin'] = pd.cut(titanic['age'], bins=4, labels=["Child", "Young", "Adult", "Senior"])

# Create three fare bins based on quartiles and label them
titanic['fare_bin'] = pd.qcut(titanic['fare'], q=3, labels=["low", "medium", "high"])

# Create a cross-tabulation of the two new categorical columns
crosstab = pd.crosstab(titanic['age_bin'], titanic['fare_bin'])

# Print the cross-tabulation table to the console
print("Cross-tabulation table:")
print(crosstab)

# Create a heatmap of the cross-tabulation data
sns.heatmap(crosstab, annot=True, cmap="Blues", fmt="d")

# Add a title to the plot
plt.title("Age vs. Fare Bins")

# Display the plot
plt.show()
```
