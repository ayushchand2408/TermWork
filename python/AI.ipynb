```python
import pandas as pd
import matplotlib.pyplot as plt 

# Load the dataset
df = pd.read_excel(r"c:\Users\ayush\OneDrive\Documents\student_habits_performance.xlsx", engine="openpyxl")

print(df.head())
```

```python
# Create scatter plot
plt.scatter(df['student_id'], df['study_hours_per_day'])  # Changed 'data' to 'df'
plt.title("Scatter Plot") 
plt.xlabel('student_id') 
plt.ylabel('study_hours_per_day') 
plt.show()
```

```python
import seaborn as sns
```

```python

sns.regplot(x="age", y="study_hours_per_day", data=df) 
plt.title('Title using Matplotlib Function') 
plt.show()
```

```python

sns.regplot(x="age", y="study_hours_per_day", data=df)
plt.title('Study Hours vs Age')

sns.set_style("darkgrid")
sns.set_palette("pastel")

sns.boxplot(x='age', y='study_hours_per_day', data=df, hue="gender")

plt.xlabel('age')
plt.ylabel('Study Hours per Day')

plt.show()
```

```python
plt.bar(df['age'], df['study_hours_per_day'])

plt.title("Average Study Hours by age")
plt.xlabel('age')
plt.ylabel('study_per_day')

plt.show()
```

```python
sns.barplot(x='age',y='gender', data=df,hue='gender') 
plt.show() 
```

```python
plt.hist(df['age']) 
plt.title("Histogram") 
plt.show()
```

```python
plt.plot(df['age']) 
plt.plot(df['gender']) 
plt.title("Scatter Plot") 
plt.xlabel('age') 
plt.ylabel('gender') 
plt.show() 
```

```python
sns.lineplot(x='age', y='gender', data=df) 
plt.show() 
```
